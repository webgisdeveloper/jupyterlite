## Total CO2 diagram
1) Soil calcite  
Solution 1-10, average, C1-10 (mol/L_pw) * V (500000 L_pw/ha) * 44 (g/mol)/1000000 = M (tones CO2/ha)  
2) Effluent calcite  
Solution 11, C11 (mol/L_pw) * V (500000/10 L_pw/ha) * 44 (g/mol)/1000000 = M (tones CO2/ha)  
3) Effluent bicarbonate  
Solution 11, Sum C11 (mol/L_pw) * V (500000/10 L_pw/ha) * 44 (g/mol)/1000000 = M (tones CO2/ha)  
4) Total capture CO2
Effluent bicarbonate + Effluent calcite + Soil calcite

```python
# soil calcite
# select solution 1-10, average 1-10 by years"no-latex"
soil_c = (df[df['soln'].between(1,10)])[["step","years","Calcite"]].groupby("step").mean()
# convert to M (tones CO2/ha)
soil_c['Calcite'] = soil_c['Calcite']*500000*44/1000000
# rename
soil_c.rename(columns={"Calcite": "Soil_Calcite"},inplace = True)
```

```python
# Effluent calcite
# select solution 11, average 11 by years
eff_c = (df[df['soln'] == 11])[["step","years","Calcite"]].groupby('step').mean()
# convert to M (tones CO2/ha)
eff_c['Calcite'] = eff_c['Calcite']*(500000/10)*44/1000000
# rename
eff_c.rename(columns={"Calcite": "Effluent_Calcite"},inplace = True)
```

```python
# 'Effluent bicarbonate' shall use C(4) rather than Calcite
# select solution 11, average 11 by years
eff_b = (df[df['soln'] == 11])[["step","years","C(4)"]].groupby('step').mean()
# convert to M (tones CO2/ha)
eff_b['C(4)'] = eff_b['C(4)']*(500000/10)*44/1000000
eff_b['Effluent_bicarbonate']=eff_b["C(4)"].cumsum()
```

```python
eff_c["Soil_Calcite"]= soil_c["Soil_Calcite"]
eff_c["Effluent_bicarbonate"]=eff_b["Effluent_bicarbonate"]
eff_c["Total_Capture"] = eff_c['Effluent_bicarbonate'] + eff_c["Effluent_Calcite"]+eff_c["Soil_Calcite"]
```

```python
# plot everything together
fig, ax = plt.subplots()
columns = ['Effluent_Calcite','Effluent_bicarbonate','Soil_Calcite','Total_Capture']
for column in columns:
    ax.scatter(x = eff_c['years'], y= eff_c[column], label=column.replace("_"," "),s=10)
ax.set_xlabel('years')
ax.set_ylabel("Carbon captuere\n(tones CO2/ha)")
ax.legend()
ax.autoscale(tight=True)
```

## Vertical Profile in Solutes (water)

```python
# para_list for water
para_list_water = ['pH','N','Na','Ca','Mg','K','Al','Si','Fe','Mn','Sr','Ba','P','C(4)','Hfo_s','Hfo_w',
                   'm_CaX2','m_KX','m_MgX2','m_NaX','m_FeX2','m_MnX2','m_SrX2','m_AlX3','m_BaX2','m_Hfo_wH2PO4','m_Hfo_wHPO4-',
                   'Al(OH)3(a)','Calcite','SiO2(a)','Gibbsite','Goethite','Kaolinite','Fe(OH)3(a)','Pyrolusite',
                   'k_Quartz','k_Plagioclase','k_Apatite','k_Diopside_Mn','k_Diopside','k_Olivine','k_Alkali-feldspar','k_Montmorillonite','k_Ilmenite','k_Glass','k_MikeSorghum','SurfH_Ca','SurfH_Mg','SurfH_H','Hfo_PO4']
```

```python
# remove no existing symbol

def verfy_symbol(para_list):
    """verfy the symbol name"""
    not_found = []
    for symbol in para_list:
        if not symbol in column_names:
            print(f"{symbol} is not fond!")
            not_found.append(symbol)
    para_list = [x for x in para_list if not x in not_found]
    return para_list
```

```python
para_list_water = verfy_symbol(para_list_water)
```

```python
# set unit for water profile
def set_unit_water(symbol):
    if para_list_water.index(symbol) <= para_list_water.index('C(4)'):
        unit = "mg/L"
    else:
        unit = "mmol/L"
    if symbol == "pH":
        unit = ""
    return unit
```

```python
# set scale for water profile
def set_scale_water(symbol):
    if para_list_water.index(symbol) <= para_list_water.index('Hfo_w'):
        mv=mw_df.loc[mw_df['symbol']==symbol,"molecular_weight"].values[0]
        scale = 1000 * float(mv)
    else:
        scale = 1000
    if symbol == "pH":
        scale = 1.0
    return scale
```

```python
# vertical_profile_plot function
def vertical_profile_water_plot(para):
    """plot the vertical profile for a given parameter"""
    print(f'Ploting {para}')
    # mol/kgw to mg/L
    scale = set_scale_water(para)
    unit = set_unit_water(para)
    caption = mw_df.loc[mw_df['symbol']==para,"caption"].values[0]
    # three times
    years = list(df['years'].unique())
    year3 = [years[0], years[len(years)//2],years[-1]]
    # plot everything together
    fig, ax = plt.subplots()
    for year in year3:
        df_ele = df[df['years']== year][['dist_x',para]]
        df_ele = df_ele[df_ele['dist_x'] >= 0.0]
        df_ele[para]=df_ele[para] * scale
        ax.plot(df_ele[para], df_ele['dist_x'], marker='o',label=f'{year:.1f} year')
    ax.set_ylabel("Depth(m)")
    if unit == "":
        ax.set_title(f'{caption} in water')
    else:
        ax.set_title(f'{caption} ({unit}) in water')
    ax.set_xlim(left=0.0)
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.2f}"))
    ax.invert_yaxis()
    ax.legend()
    plt.show()
```

```python
# build a widget
para_widget = widgets.Dropdown(options=para_list_water, description='Select a parameter to plot: ', style= {'description_width': 'initial'})
widgets.interact(vertical_profile_water_plot, para = para_widget);
```

## Heat Map
For a parameter y = depth (reversed), x = years

```python
# add si_ to para_list
para_list_heatmap = ['pH','N','Na','Ca','Mg','K','Al','Si','Fe','Mn','Sr','Ba','P','C(4)','Hfo_s','Hfo_w',
                     'm_CaX2','m_KX','m_MgX2','m_NaX','m_FeX2','m_MnX2','m_SrX2','m_AlX3','m_BaX2','m_Hfo_wH2PO4','m_Hfo_wHPO4-',
                     'Al(OH)3(a)','Calcite','SiO2(a)','Gibbsite','Goethite','Kaolinite','Fe(OH)3(a)','Pyrolusite',
                     'k_Quartz','k_Plagioclase','k_Apatite','k_Diopside_Mn','k_Diopside','k_Olivine','k_Alkali-feldspar','k_Montmorillonite','k_Ilmenite','k_Glass',
                     'si_Al(OH)3(a)','si_Calcite','si_SiO2(a)','si_Gibbsite','si_Fe(OH)3(a)','si_Pyrolusite','si_Kaolinite','si_Quartz','si_Plagioclase','si_Apatite',
                     'si_Diopside_Mn','si_Diopside','si_Olivine','si_Alkali-feldspar','si_Montmorillonite','si_Ilmenite','si_Glass']
```

```python
para_list_heatmap = verfy_symbol(para_list_heatmap)
```

```python
# set unit for heatmap
def set_unit_heatmap(symbol):
    if para_list_heatmap.index(symbol) <= para_list_heatmap.index('C(4)'):
        unit = "mg/L"
    elif para_list_heatmap.index(symbol) > para_list_heatmap.index('C(4)') and para_list_heatmap.index(symbol) <= para_list_heatmap.index('m_Hfo_wHPO4-'):
        unit = "mmol/L"
    else:
        #unit = "mg/L_soil"
        unit = "mg/kg_soil"
    if symbol == "pH" or "si_" in symbol:
        unit = ""
    return unit
```

```python
# set scale for heatmap
def set_scale_heatmap(symbol):
    """ set scale by unit"""
    effective_porosity = 0.2
    unit = set_unit_heatmap(symbol)
    mv=mw_df.loc[mw_df['symbol']==symbol,"molecular_weight"].values[0]
    if unit == "":
        scale = 1.0
    elif unit == "mg/L":
        scale = 1000 * float(mv)
    elif unit == "mmol/L":
        scale = 1000
    elif unit == "mg/kg_soil":
        soil_density = 1.7
        scale = effective_porosity * mv / soil_density
    else:
        scale = 1.0
    return scale
```

```python
def heatmap_plot(para,aspectfix=False):
    """heatmap plot"""
    print(f'plotting {para}')
    unit = set_unit_heatmap(para)
    scale = set_scale_heatmap(para)
    #print(scale)
    caption = mw_df.loc[mw_df['symbol']==para,"caption"].values[0]
    # collect the data
    df_hm = df[['years','dist_x',para]]
    df_hm = df_hm[df_hm['dist_x'] >= 0.0]
    year_unique = df_hm['years'].unique().tolist()
    depth_unique = df_hm['dist_x'].unique().tolist()
    # depth is row, year is column
    row_m = len(depth_unique)
    column_n = len(year_unique)
    # need to sort by two depth and years first
    df_hm.sort_values(by=['dist_x',"years"], inplace = True)
    # reshape element as a array row_n, column_ninline
    para_array = (scale*df_hm[para].values).reshape(row_m, column_n)
    extent = [min(year_unique), max(year_unique), max(depth_unique), min(depth_unique)]
    fig, ax = plt.subplots(figsize=(7,7))
    pos = ax.imshow(para_array,extent=extent,interpolation="none")
    fig.colorbar(pos, ax=ax, shrink=0.75)
    if aspectfix:
        ax.set_aspect(1)
    else:
        ax.set_aspect(column_n/row_m)
    if unit == "":
        ax.set_title(f'{caption}')
    else:
        ax.set_title(f'{caption} ({unit})')
    ax.set_ylabel("Depths (m)")
    ax.set_xlabel("years")
    plt.show()
```

```python
# build a widget
heatmap_widget = widgets.Dropdown(options=para_list_heatmap, description='Select a parameter to plot: ', style= {'description_width': 'initial'})
aspect_widget =  widgets.Checkbox(value=False, description='Equal Aspect Ratio')

widgets.interact(heatmap_plot, para = heatmap_widget, aspectfix = aspect_widget);
```
